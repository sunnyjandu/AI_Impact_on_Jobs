# Section 1

# Extract

1. Import libraries necessary for data analysis and visualisation tasks.
2. The pandas library will be utilised for data analysis. 

```python
# importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# using pandas library to read the csv data in
df_full = pd.read_csv('../Dataset/Raw/AI_Impact_on_Jobs_2030.csv')
```

# Section 2

# Transform

1. Explore the data and limited to the first 500 rows to make analysis more manageable.
2. Check for missing values and drop rows with missing values.
3. Remove duplicate data.
4. Standardise formats.

```python
# Explore the data
df = df_full.head(500)  # Limit to 500 rows for manageability
print("Shape:", df.shape)
print("Head:")
display(df.head())
print("Info:")
df.info()
print("Describe:")
display(df.describe())

# Check for missing values
print("Missing values:")
print(df.isnull().sum())

# Handle missing values (drop rows with missing values as an example)
df = df.dropna()

# Remove duplicates
df = df.drop_duplicates()

# Standardise formats (e.g., lowercase string columns)
df = df.apply(lambda x: x.str.lower() if x.dtype == 'object' else x)

print("Data cleaned. Shape after cleaning:", df.shape)
```

```python
# Check counts for key columns
print("Risk_Category counts:")
print(df['Risk_Category'].value_counts())

print("\nEducation_Level counts:")
print(df['Education_Level'].value_counts())

print("\nJob_Title counts (top 10):")
print(df['Job_Title'].value_counts().head(10))
```

# Section 3

# Load

1. Load the data to the "Cleaned" folder as CSV file. 

```python
# Save the cleaned data to the Cleaned folder
df.to_csv('../Dataset/Cleaned/cleaned_data.csv', index=False)
print("Cleaned data saved to ../Dataset/Cleaned/cleaned_data.csv")
```

```python
# Load and display the cleaned data as a table
df_cleaned = pd.read_csv('../Dataset/Cleaned/cleaned_data.csv')
display(df_cleaned.head(10))  # Display first 10 rows as a table
```

```python
# Hypothesis 1 ETL: Prepare data for entry-level or graduate jobs at higher risk of automation

# Filter for entry-level jobs (low experience, e.g., <= 3 years) or graduate jobs (higher education: Master's or PhD)
df_h1 = df[(df['Years_Experience'] <= 3) | (df['Education_Level'].isin(['master\'s', 'phd']))]

# Separate into routine/manual tasks (High risk) and cognitive/creative roles (Low risk)
df_routine = df_h1[df_h1['Risk_Category'] == 'High']
df_cognitive = df_h1[df_h1['Risk_Category'] == 'Low']

# Save the filtered datasets for analysis
df_h1.to_csv('../Dataset/Cleaned/hypothesis1_data.csv', index=False)
df_routine.to_csv('../Dataset/Cleaned/hypothesis1_routine.csv', index=False)
df_cognitive.to_csv('../Dataset/Cleaned/hypothesis1_cognitive.csv', index=False)

print("Hypothesis 1 data prepared and saved.")
print(f"Total jobs in H1: {len(df_h1)}")
print(f"Routine/manual jobs: {len(df_routine)}")
print(f"Cognitive/creative jobs: {len(df_cognitive)}")
```

```python
# Hypothesis 2 ETL: Prepare summary for jobs by education level and average automation risk

# Group by education level and calculate average automation risk
df_h2 = df.groupby('Education_Level')['Automation_Probability_2030'].mean().reset_index()
df_h2.rename(columns={'Automation_Probability_2030': 'Average_Risk'}, inplace=True)

# Save the summary
df_h2.to_csv('../Dataset/Cleaned/hypothesis2_summary.csv', index=False)
print("Hypothesis 2 summary saved.")
print(df_h2)
```

```python
# Hypothesis 3 ETL: Prepare correlations between skills and automation risk

# Identify skill columns
skill_cols = [f'Skill_{i}' for i in range(1, 11)]

# Calculate correlations with automation risk
correlations = df[skill_cols + ['Automation_Probability_2030']].corr()['Automation_Probability_2030'].drop('Automation_Probability_2030')
df_h3 = correlations.reset_index()
df_h3.columns = ['Skill', 'Correlation_with_Risk']

# Save the correlations
df_h3.to_csv('../Dataset/Cleaned/hypothesis3_correlations.csv', index=False)
print("Hypothesis 3 correlations saved.")
print(df_h3)
```
