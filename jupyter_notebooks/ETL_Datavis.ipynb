# Extract

1. Import libraries necessary for data analysis and visualisation tasks.
2. The pandas library will be utilised for data analysis. 

```python
# importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "browser"

# using pandas library to read the csv data in
df_full = pd.read_csv('../Dataset/Raw/AI_Impact_on_Jobs_2030.csv')
```

# Transform

1. Explore the data and limited to the first 3000 rows to make analysis more manageable.
2. Check for missing values and drop rows with missing values.
3. Remove duplicate data.
4. Standardise formats.

```python
# Explore the data
df = df_full.head(3000)  # Limit to 300 rows for manageability
print("Shape:", df.shape)
print("Head:")
display(df.head())
print("Info:")
df.info()
print("Describe:")
display(df.describe())

# Check for missing values
print("Missing values:")
print(df.isnull().sum())

# Handle missing values (drop rows with missing values as an example)
df = df.dropna()

# Remove duplicates
df = df.drop_duplicates()

# Standardise formats (e.g., lowercase string columns)
df = df.apply(lambda x: x.str.lower() if x.dtype == 'object' else x)

print("Data cleaned. Shape after cleaning:", df.shape)
```

# Data Visualisation

## Hypothesis 1
Entry-level or graduate jobs, which often involve routine or manual tasks, are at higher risk of AI-driven automation by 2030.

```python
# Filter and Copy the Data
df_h1 = df[
    (df['Years_Experience'] <= 5) |
    (df['Education_Level'].isin(["master's", "phd"]))
].copy()

# Standardise Risk_Category 
df_h1.loc[:, 'Risk_Category'] = df_h1['Risk_Category'].str.lower()

# Separate High-risk (routine/manual) and Low-risk (cognitive/creative) jobs
df_routine = df_h1[df_h1['Risk_Category'] == 'high'].copy()
df_cognitive = df_h1[df_h1['Risk_Category'] == 'low'].copy()

print("Hypothesis 1 data prepared and saved.")
print(f"Total jobs in H1: {len(df_h1)}")
print(f"Routine/manual jobs (High risk): {len(df_routine)}")
print(f"Cognitive/creative jobs (Low risk): {len(df_cognitive)}")

# Display actual job titles for high-risk (routine/manual) jobs
print("\nHigh-risk (Routine/Manual) jobs:")
print(df_routine[['Job_Title', 'Risk_Category']].reset_index(drop=True))

# Display actual job titles for low-risk (Cognitive/Creative) jobs
print("\nLow-risk (Cognitive/Creative) jobs:")
print(df_cognitive[['Job_Title', 'Risk_Category']].reset_index(drop=True))

# Count risk categories
risk_counts = df_h1['Risk_Category'].value_counts()

# --- Matplotlib visualization ---
plt.figure(figsize=(8, 5))
plt.bar(
    risk_counts.index,
    risk_counts.values,
    color=['red', 'orange', 'green'][:len(risk_counts)]
)
plt.title('Automation Risk Distribution (Hypothesis 1 - Matplotlib)')
plt.xlabel('Risk Category')
plt.ylabel('Number of Jobs')
plt.tight_layout()
plt.show()

# --- Seaborn visualization ---
plt.figure(figsize=(8, 5))
sns.barplot(
    x=risk_counts.index,
    y=risk_counts.values,
    hue=risk_counts.index,    
    palette='Set2',
    dodge=False
)
plt.title('Automation Risk Distribution (Hypothesis 1 - Seaborn)')
plt.xlabel('Risk Category')
plt.ylabel('Number of Jobs')
plt.tight_layout()
plt.show()

# --- Plotly visualization ---
risk_df = risk_counts.reset_index()
risk_df.columns = ['Risk_Category', 'Count']

fig = px.bar(
    risk_df,
    x='Risk_Category',
    y='Count',
    color='Risk_Category',
    title='Automation Risk Distribution (Hypothesis 1 - Plotly)'
)
fig.show()
```

## Hypothesis 2
Jobs that require higher education are less likely to be automated.

```python
# Normalize education level 
df['Education_Level'] = df['Education_Level'].str.lower()

# Map education levels to numeric values for correlation 
edu_numeric = {'high school': 1, 'bachelor': 2, 'master': 3, 'phd': 4}
df['Edu_Num'] = df['Education_Level'].map(edu_numeric)

# Aggregate by job to avoid duplicates
df_jobs = df.groupby(['Job_Title', 'Education_Level', 'Edu_Num'], as_index=False)['Automation_Probability_2030'].mean()

# Calculate average automation risk per education level 
df_avg = df_jobs.groupby(['Education_Level', 'Edu_Num'], as_index=False)['Automation_Probability_2030'].mean()
df_avg.rename(columns={'Automation_Probability_2030':'Average_Risk'}, inplace=True)
df_avg = df_avg.sort_values('Edu_Num')  # Ensure proper order

# Line plot to show trend 
plt.figure(figsize=(8,5))
plt.plot(df_avg['Edu_Num'], df_avg['Average_Risk'], marker='o', color='red', linewidth=2)
plt.xticks(df_avg['Edu_Num'], df_avg['Education_Level'])  # Label ticks with education levels
plt.xlabel('Education Level')
plt.ylabel('Average Automation Risk')
plt.title('Average Automation Risk by Education Level')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Seaborn barplot with trendline
plt.figure(figsize=(8,5))
sns.barplot(data=df_avg, x='Education_Level', y='Average_Risk', color='lightblue')
sns.lineplot(data=df_avg, x='Education_Level', y='Average_Risk', marker='o', color='red')
plt.xlabel('Education Level')
plt.ylabel('Average Automation Risk')
plt.title('Average Automation Risk by Education Level')
plt.tight_layout()
plt.show()

# Correlation between education level and automation risk
correlation = df_jobs[['Edu_Num','Automation_Probability_2030']].corr().iloc[0,1]
print(f"Correlation between education level and automation risk: {correlation:.2f}")
```

## Hypothesis 3
Jobs that need advanced digital or thinking skills (like creativity, communication, or technical skills) are safer from AI automation.

```python
# Identify skill columns
skill_cols = [f'Skill_{i}' for i in range(1, 11)]

# Calculate correlations with automation risk
correlations = df[skill_cols + ['Automation_Probability_2030']].corr()['Automation_Probability_2030'].drop('Automation_Probability_2030')

# Convert to DataFrame for plotting
df_skills_corr = correlations.reset_index()
df_skills_corr.columns = ['Skill', 'Correlation_with_Risk']

# Sort by correlation (optional, makes visualization easier to read)
df_skills_corr = df_skills_corr.sort_values(by='Correlation_with_Risk', ascending=True)

# Matplotlib Line Plot
plt.figure(figsize=(10, 6))
plt.plot(
    df_skills_corr['Skill'],
    df_skills_corr['Correlation_with_Risk'],
    marker='o',
    linestyle='-',
    color='seagreen'
)
plt.title('Correlation Between Skills and Automation Risk')
plt.xlabel('Skill')
plt.ylabel('Correlation Coefficient')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Seaborn Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_skills_corr,
    x='Skill',
    y='Correlation_with_Risk',
    hue='Correlation_with_Risk', 
    palette='coolwarm',
    s=100,
    legend=False
)
plt.title('Correlation Between Skills and Automation Risk')
plt.xlabel('Skill')
plt.ylabel('Correlation Coefficient')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
