# Extract

1. Import libraries necessary for data analysis and visualisation tasks.

```python
# importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "browser"

# using pandas library to read the csv data in
df_full = pd.read_csv('../Dataset/Raw/AI_Impact_on_Jobs_2030.csv')
```

# Transform

1. Explore the data and limited to the first 500 rows to make analysis more manageable.
2. Check for missing values and drop rows with missing values.
3. Remove duplicate data.
4. Standardise formats.

```python
# Explore the data
df = df_full.head(500)  # Limit to 500 rows for manageability
print("Shape:", df.shape)
print("Head:")
display(df.head())
print("Info:")
df.info()
print("Describe:")
display(df.describe())

# Check for missing values
print("Missing values:")
print(df.isnull().sum())

# Handle missing values (drop rows with missing values as an example)
df = df.dropna()

# Remove duplicates
df = df.drop_duplicates()

# Standardise formats (e.g., lowercase string columns)
df = df.apply(lambda x: x.str.lower() if x.dtype == 'object' else x)

print("Data cleaned. Shape after cleaning:", df.shape)
```

```python
# Check counts for key columns
print("Risk_Category counts:")
print(df['Risk_Category'].value_counts())

print("\nEducation_Level counts:")
print(df['Education_Level'].value_counts())

print("\nJob_Title counts (top 10):")
print(df['Job_Title'].value_counts().head(10))
```

# Data Visualisation

## Hypothesis 1
Entry-level or graduate jobs, which often involve routine or manual tasks, are at higher risk of AI-driven automation by 2030.

```python
# Create a copy to avoid SettingWithCopyWarning
df_h1 = df[
    (df['Years_Experience'] <= 5) |
    (df['Education_Level'].isin(["master's", "phd"]))
].copy()

# Lowercase the Risk_Category column safely
df_h1.loc[:, 'Risk_Category'] = df_h1['Risk_Category'].str.lower()

# Separate High-risk (routine/manual) and Low-risk (cognitive/creative) jobs
df_routine = df_h1[df_h1['Risk_Category'] == 'high'].copy()
df_cognitive = df_h1[df_h1['Risk_Category'] == 'low'].copy()

print("Hypothesis 1 data prepared and saved.")
print(f"Total jobs in H1: {len(df_h1)}")
print(f"Routine/manual jobs (High risk): {len(df_routine)}")
print(f"Cognitive/creative jobs (Low risk): {len(df_cognitive)}")

# Risk distribution
risk_counts = df_h1['Risk_Category'].value_counts()

# --- Matplotlib visualization ---
plt.figure(figsize=(8, 5))
plt.bar(
    risk_counts.index,
    risk_counts.values,
    color=['red', 'orange', 'green'][:len(risk_counts)]
)
plt.title('Automation Risk Distribution (Hypothesis 1 - Matplotlib)')
plt.xlabel('Risk Category')
plt.ylabel('Number of Jobs')
plt.tight_layout()
plt.show()

# --- Seaborn visualization ---
plt.figure(figsize=(8, 5))
sns.barplot(
    x=risk_counts.index,
    y=risk_counts.values,
    hue=risk_counts.index,    
    palette='Set2',
    dodge=False
)
plt.title('Automation Risk Distribution (Hypothesis 1 - Seaborn)')
plt.xlabel('Risk Category')
plt.ylabel('Number of Jobs')
plt.tight_layout()
plt.show()

# --- Plotly visualization ---
risk_df = risk_counts.reset_index()
risk_df.columns = ['Risk_Category', 'Count']

fig = px.bar(
    risk_df,
    x='Risk_Category',
    y='Count',
    color='Risk_Category',
    title='Automation Risk Distribution (Hypothesis 1 - Plotly)'
)
fig.show()
```

## Hypothesis 2
Jobs that require higher education are less likely to be automated.

```python
# Normalize education level text
df['Education_Level'] = df['Education_Level'].str.lower()

df_h2 = (
    df
    .groupby('Education_Level', as_index=False)['Automation_Probability_2030']
    .mean()
)

df_h2.rename(
    columns={'Automation_Probability_2030': 'Average_Risk'},
    inplace=True
)


# Matplotlib visualization
plt.figure(figsize=(8, 5))
plt.bar(
    df_h2['Education_Level'],
    df_h2['Average_Risk'],
    color='skyblue'
)
plt.title('Average Automation Risk by Education Level (Hypothesis 2 - Matplotlib)')
plt.xlabel('Education Level')
plt.ylabel('Average Automation Probability')
plt.tight_layout()
plt.show()


# Seaborn visualization (FIXED warning)
plt.figure(figsize=(8, 5))
sns.barplot(
    data=df_h2,
    x='Education_Level',
    y='Average_Risk',
    hue='Education_Level',   # FIX
    palette='Blues',
    legend=False
)
plt.title('Average Automation Risk by Education Level (Hypothesis 2 - Seaborn)')
plt.xlabel('Education Level')
plt.ylabel('Average Automation Probability')
plt.tight_layout()
plt.show()

# Plotly visualization 
fig = px.bar(
    df_h2,
    x='Education_Level',
    y='Average_Risk',
    color='Education_Level',
    title='Average Automation Risk by Education Level (Hypothesis 2 - Plotly)'
)

fig.show()
```

## Hypothesis 3
Jobs that need advanced digital or thinking skills (like creativity, communication, or technical skills) are safer from AI automation.

```python
# Identify skill columns
skill_cols = [f'Skill_{i}' for i in range(1, 11)]

# Calculate correlations
correlations = (
    df[skill_cols + ['Automation_Probability_2030']]
    .corr()['Automation_Probability_2030']
    .drop('Automation_Probability_2030')
)

df_h3 = correlations.reset_index()
df_h3.columns = ['Skill', 'Correlation_with_Risk']


# Matplotlib visualization
plt.figure(figsize=(10, 6))
plt.plot(
    df_h3['Skill'],
    df_h3['Correlation_with_Risk'],
    marker='o',
    linestyle='-',
    color='seagreen'
)
plt.title('Correlation between Skills and Automation Risk (Hypothesis 3 - Matplotlib)')
plt.xlabel('Skill')
plt.ylabel('Correlation Coefficient')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Seaborn visualization (FIXED warning)
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_h3,
    x='Skill',
    y='Correlation_with_Risk',
    hue='Skill',        # FIX: enables palette
    palette='viridis',
    s=100,
    legend=False
)
plt.title('Correlation between Skills and Automation Risk (Hypothesis 3 - Seaborn)')
plt.xlabel('Skill')
plt.ylabel('Correlation Coefficient')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# Plotly visualization (INTERACTIVE)
fig = px.bar(
    df_h3,
    x='Skill',
    y='Correlation_with_Risk',
    color='Correlation_with_Risk',
    color_continuous_scale='RdYlGn_r',
    title='Correlation between Skills and Automation Risk (Hypothesis 3 - Plotly)'
)

fig.show()
```
